```python
import pandas as pd # 판다스
import plotly.graph_objs as go # 예쁜 그래프 그리기
import plotly.offline as py # 예쁜 그래프 그리기
from fbprophet import Prophet # 프로펫
from fbprophet.plot import plot_plotly, add_changepoints_to_plot # 프로펫
py.init_notebook_mode
```

```python
df = pd.read_csv('/content/전국감염병발생수01-20.csv', encoding='cp949')
df.head()
```

```python
df = df.rename(columns={'시간':'time', '감염병발생수':'confirmed'})
df.tail()
```

```python
df["time"] = df["time"].str.replace(pat=r'년', repl=r'-', regex=True)
```

```python
df["time"] = df["time"].str.replace(pat=r'월', repl=r'', regex=True)
```

```python
df.head()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df.time,
        y=df.confirmed,
        name='Confirmed in Korea'
    )
)
```

```python
df_prophet = df.rename(columns={
    'time':'ds',
    'confirmed':'y'
})

df_prophet.head()
```

```python
m = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='additive'
)

m.fit(df_prophet)

future = m.make_future_dataframe(periods=12)

future.tail(12)
```

```python
forecast = m.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12)
```

```python
fig = plot_plotly(m, forecast)
py.iplot(fig) # 까만점이 실제 데이터이다. # 파란선은 예측값 # 하늘색은 오차 범위
```

```python
m = Prophet(
    changepoint_prior_scale=0.98, # 숫자가 커질수록 모델이 유연해진다.
    changepoint_range=0.2, # 앞부분의 얼만큼을 보고 예측할것이냐 기본값 0.8.
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='additive'
)

m.fit(df_prophet)

future = m.make_future_dataframe(periods=12)
forecast = m.predict(future)

fig = plot_plotly(m, forecast)
py.iplot(fig)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12)
```

```python
import pandas as pd # 판다스
df = pd.read_csv('/content/전국월별평균기온73-20.csv', encoding='cp949') # 데이터 로드
```

```python
df.head()
```

```python
df = df.rename(columns={'시간':'time', '평균기온':'C'})
df.tail() # 이름 바꿔줬다
```

```python
df["time"] = df["time"].str.replace(pat=r'년', repl=r'-', regex=True) # 한글 지워줬다
```

```python
df["time"] = df["time"].str.replace(pat=r'월', repl=r'', regex=True) # 한글 지워줬다
```

```python
df.head()
```

```python
from matplotlib import pyplot # 맷플롯립으로 온도를 그려봤다
df.plot()
pyplot.show() # 이쁘다
```

```python
df.columns = ['ds', 'y']
df['ds'] = pd.to_datetime(df['ds']) # 프로펫은 변수 이름을 ds, y 로 맞춰줘야 한단다
```

```python
from fbprophet import Prophet # 프로펫
model = Prophet() # 이제부터 너는 모델이다
```

```python
model.fit(df) # 학습
```

```python
last_1year = list() # 마지막 년도인 2020년의 12개의 달을 모으려고 한다 
for i in range(1, 13):
  last_1year.append(['2020-%02d' % i])
last_1year = pd.DataFrame(last_1year, columns = ['ds'])
last_1year['ds'] = pd.to_datetime(last_1year['ds']) # 남에 꺼 따라했다

# 미래나 과거를 예측해보기 전 학습으로 사용한 데이터를 잘 예측하는지 확인해야 한단다
# in_sample_forecast라고 한다는데, 즉 train_set을 잘 예측하냐 보는거다
```

```python
forecast = model.predict(last_1year) # 모은 2020년을 예측해보라 하고 forecast에 저장했다
```

```python
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()) 
# 뭐가 많은데 4가지 컬럼만 보겠다
```

```python
model.plot(forecast)
pyplot.show()
# 프로펫은 내장함수 plot()가 있어서 좋다
# 검은점은 실제 데이터고 파란선이 예측값인데 엄청 잘 맞춘듯
```

```python
last_1year1 = list() # 이번엔 임의의 날짜 2021년 12달을 만든다
for i in range(1, 13):
  last_1year1.append(['2021-%02d' % i])
last_1year1 = pd.DataFrame(last_1year1, columns = ['ds'])
last_1year1['ds'] = pd.to_datetime(last_1year1['ds'])

# out_of_sample_forecast라고 한다는데, 그냥 21년 예측하는거다
```

```python
forecast1 = model.predict(last_1year1)
model.plot(forecast1)
pyplot.show()
# 대충 잘 맞춘듯, 점이 많으니까 그냥 용과같아서 이쁘다
```

```python
# 이번에는 다르게 마지막 1년을 제외한 값으로 학습을 하고 마지막 1년을 예측한 뒤 
# 실제 마지막 1년과 비교하려고 한다 

# 마지막 12개월 제외시키기 
train = df.drop(df.index[-12:]) # train에서 뒤에 12개 제외
y_true = df['y'][-12:].values # 제외한 값은 y_true로 저장해둔다

# 모델 생성 후 학습
model = Prophet()
model.fit(train) # 12개가 빠진 train으로 학습

# 가상의 마지막 12개월 생성
last_1year = list()
for i in range(1, 13):
  last_1year.append(['2020-%02d' % i])
last_1year = pd.DataFrame(last_1year, columns = ['ds'])
last_1year['ds'] = pd.to_datetime(last_1year['ds']) # 실제 값이 들은게 아니다 !

# 예측, 비교하기 그리고 MAE를 살펴보기
forecast = model.predict(last_1year)
y_pred = forecast['yhat'].values
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true, y_pred)
print('MAE: %.3f' % mae)
```

```python
pyplot.plot(y_true, label='Actual')
pyplot.plot(y_pred, label='Predicted')
pyplot.legend()
pyplot.show()
# MAE 1.461 잘 맞춘 듯 하다
```

```python
!pip install pandas_datareader
!pip install pmdarima
import os

import pandas as pd
import pandas_datareader as pdr

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
import matplotlib
plt.style.use('seaborn-whitegrid')

import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima.arima import auto_arima

import seaborn as sns
plt.style.use('seaborn-whitegrid')
#sns.get.style('white')

import itertools
```

```python
data = pd.read_csv('/content/전국감염병발생수01-20.csv', encoding='cp949')
data.head()
```

```python
data['시간'] = data['시간'].str.replace(pat=r'년', repl=r'-', regex=True)
```

```python
data['시간'] = data['시간'].str.replace(pat=r'월', repl=r'', regex=True)
```

```python
data = data.rename(columns={'시간':'month', '감염병발생수':'confirmed'})
```

```python
data['month'] = pd.to_datetime(data['month'])
```

```python
data = data.set_index('month')
```

```python
data
```

```python
fig = data.plot()
```

```python
decompostion = sm.tsa.seasonal_decompose(data['confirmed'], model='additive')
fig = decompostion.plot()
fig.set_size_inches(10,10)
plt.show()
# Observed: Observed data
# Trend: The estimated trend component 추세
# Seasonal: The estimated seasonal component 계절성
# Residual: The estimated residuals 잔차
```

```python
train_data, test_data = train_test_split(data, test_size=0.2, shuffle=False) # 셔플없이 뒤에 20퍼센트가 test가 되게 했다.
```

```python
# Raw Data의 acf, pacf 플롯을 그린다.
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle('Raw Data')
sm.graphics.tsa.plot_acf(train_data.values.squeeze(), lags=30, ax=ax[0])
sm.graphics.tsa.plot_pacf(train_data.values.squeeze(), lags=30, ax=ax[1]);
```

```python
# 차분하다.
diff_train_data = train_data.copy()
diff_train_data = diff_train_data['confirmed'].diff()
diff_train_data = diff_train_data.dropna()
print('######## Raw Data ########')
print(train_data)
print('#### Differences Data ####')
print(diff_train_data)
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(211)
plt.plot(train_data['confirmed'])
plt.legend(['Raw Data (Nonstationary)']) # 실제 데이터
plt.subplot(212)
plt.plot(diff_train_data, 'orange')
plt.legend(['Differenced Data (Stationary)']) # 1차 차분
plt.show()
```

```python
# 1차차분 후의 acf, pacf 플롯을 그린다.
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle('Differenced Data')
sm.graphics.tsa.plot_acf(diff_train_data.values.squeeze(), lags=30, ax=ax[0])
sm.graphics.tsa.plot_pacf(diff_train_data.values.squeeze(), lags=30, ax=ax[1]);
```

ARIMA 모델링

```python
model = ARIMA(train_data.values, order=(1, 1, 0))
model_fit = model.fit()
model_fit.summary()
```

```python
print('Examples of parameter combinations for Seasonal ARIMA..')
p = range(0, 3)
d = range(1, 2)
q = range(0, 3)
pdq = list(itertools.product(p, d, q))

aic=[]
for i in pdq:
  model = ARIMA(train_data.values, order=(i))
  model_fit = model.fit()
  print(f'ARIMA: {i} >> AIC : {round(model_fit.aic,2)}')
  aic.append(round(model_fit.aic,2))
```

```python
optimal = [(pdq[i], j) for i, j in enumerate(aic) if j == min(aic)]
optimal
```

```python
model_opt = ARIMA(train_data.values, order=optimal[0][0])
model_opt_fit = model_opt.fit()

model_opt_fit.summary()
```

```python
prediction = model_opt_fit.forecast(len(test_data))
predicted_value = prediction[0]
predicted_ub = prediction[2][:,0]
predicted_lb = prediction[2][:,1]
predict_index = list(test_data.index)
r2 = r2_score(test_data, predicted_value)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(predict_index, predicted_value, label = 'Prediction')
data.plot(ax=ax);
ax.vlines('2017-08-01', 0, 1000, linestyle='--', color='r', label='Start to Forecast');
ax.fill_between(predict_index, predicted_lb, predicted_ub, color = 'k', alpha = 0.1, label = '0.95 prediction interval')
ax.legend(loc='upper left')
plt.suptitle(f'ARIMA {optimal[0][0]} Prediction Results (r2_score: {round(r2,2)})')
plt.show()
```

SARIMA 모델링

```python
print('Examples of paramete combinations for Seasonal ARIMA..')
p = range(0, 3)
d = range(1, 2)
q = range(0, 3)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

aic = []
params = []
for i in pdq:
  for j in seasonal_pdq:
    try:
      model = SARIMAX(train_data.values, order=(i), seasonal_order = (j))
      model_fit = model.fit()
      print(f'SARIMA: {i}{j} >> AIC : {round(model_fit.aic,2)}')
      params.append((i, j))
    except:
      continue
```

```python
optimal = [(params[i], j) for i, j in enumerate(aic) if j == min(aic)]
optimal # 값이 안나오는데 SARIMA: (1, 1, 2)(2, 1, 2, 12) >> AIC : 2971.99 이게 최저인것같다.
```

```python
optimal = [(((1, 1, 2), (2, 1, 2, 12)), 2971.99)] # 손으로 넣어줬다.
```

```python
model_opt = SARIMAX(train_data.values, order=optimal[0][0][0], seasonal_order=optimal[0][0][1])
model_opt_fit = model_opt.fit()

model_opt_fit.summary()
```

```python
prediction = model_opt_fit.get_forecast(len(test_data))
predicted_value = prediction.predicted_mean
predicted_ub = prediction.conf_int()[:,0]
predicted_lb = prediction.conf_int()[:,1]
predict_index = list(test_data.index)
r2 = r2_score(test_data, predicted_value)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(predict_index, predicted_value, label = 'Prediction')
data.plot(ax=ax);
ax.vlines('2017-08-01', 0, 700, linestyle='--', color='r', label='Start to Forecast');
ax.fill_between(predict_index, predicted_lb, predicted_ub, color = 'k', alpha = 0.1, label = '0.95 prediction interval')
ax.legend(loc='upper left')
plt.suptitle(f'SARIMA {optimal[0][0][0]},{optimal[0][0][1]} Prediction Results (r2_score: {round(r2,2)})')
plt.show()
```
