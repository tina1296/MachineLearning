```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fbprophet import Prophet

%matplotlib inline


df = pd.read_excel("/Users/seojunebeom/Desktop/대학교/2022 1학기/기계학습/전처리-WTI유가/2.xls")
```

```python
#fbprophet 사용 형식에 맞추기

df_tmp = pd.DataFrame({"ds":df["날짜"], "y":df["종가"]})
df_target = df_tmp[:-1]	# 끝에 NaN 값을 제거함

df = df_target.copy()
df["ds"] = pd.to_datetime(df_target["ds"], format="%b %d, %Y")

# .info()로 확인하면 object이기 때문에 숫자로 바꿔줌
df["y"] = df["y"].astype("float")
```

```python
#prophet 적용 & 예측

m = Prophet(yearly_seasonality=True, daily_seasonality=True)
m.fit(df);

future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)
forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail()
```

```python
#시각화

plt.figure(figsize=(12,6))
plt.plot(df["ds"], df["y"], label="real")
plt.grid(True)
plt.show()
```

```python
m.plot_components(forecast);
```
